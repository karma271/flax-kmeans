```python
from pathlib import Path
import os
import sys

if Path("/content/flax-kmeans").exists():
    os.chdir("/content/flax-kmeans")

REPO_ROOT = Path.cwd()
if not (REPO_ROOT / "src").exists():
    raise RuntimeError("Run this notebook from the flax-kmeans repo root.")

if str(REPO_ROOT) not in sys.path:
    sys.path.insert(0, str(REPO_ROOT))

print("Repo root:", REPO_ROOT)
```

```python
from pathlib import Path

import jax
import pandas as pd

from src.eval.notebook_harness import load_benchmark_matrix, run_matrix_for_implementation

print("JAX backend:", jax.default_backend())

MATRIX_PATH = Path("configs/experiments/benchmark_matrix.yaml")
matrix = load_benchmark_matrix(MATRIX_PATH)
RESULTS_ROOT = Path("results") / matrix.exp_name

records = run_matrix_for_implementation(
    matrix_path=MATRIX_PATH,
    implementation="jax_flash_kmeans",
    target_device="tpu",
    results_root=RESULTS_ROOT,
    overwrite_existing=False,
)

runs_df = pd.DataFrame(records)
runs_df.head()
```

```python
print("Total rows:", len(runs_df))
print(runs_df["status"].value_counts(dropna=False))
print("Artifacts root:", RESULTS_ROOT)
```
